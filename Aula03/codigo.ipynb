```python
import pandas as pd #importa a biblioteca pandas
import numpy as np #importa a biblioteca numpy
import matplotlib.pyplot as plt #importa a biblioteca matplotlib

Dados_x= np.linspace(0,10,10) # gera dados em x de 10
# em 10 valores
Dados_y = Dados_x * Dados_x # multiplica os dados de x 
# e armazena em y

plt.figure(figsize=(10,8),facecolor='gray') #Cria uma figura
Ax = plt.axes() # cria os eixos para o grafico
Ax.set_facecolor('white')
plt.plot(Dados_x,Dados_y,
         color='red', marker= 'o',
         linestyle='dashed',
         markeredgecolor='black') # plota o gráfico em x e y
plt.xlabel("Eixo x", fontsize=16)
plt.ylabel("Eixo y", fontsize=16)
plt.title("Gráfico simples com Matplotlib", fontsize=20)
plt.show()

```

```python
plt.rcParams.update({'font.size':15}) # função que permite ajustar a figura
plt.figure(figsize=(12,12)) #define um tamanho para a figura 
linhas =2 # cria 2 linhas
colunas =2 # cria 2 colunas
plt.subplot(linhas,colunas,1) #subplot para plotar graficos 
# um ao lado do outro
plt.plot(Dados_x,'r--', label='Dados_x') # label rotulo para os dados
plt.legend(loc = 'best') # cria legenda para o gráfico
plt.xlabel("Eixo x") # Eixo x
plt.ylabel("Eixo y") # Eixo y
plt.title("Gráfico 1") # titulo do gráfico
plt.subplot(linhas,colunas,2)
plt.plot(Dados_x,Dados_y, 'b--', label='Dados x \n e y') # plota o gráfico alterando 
# a cor da linha
plt.xlabel("Eixo x") # Eixo x
plt.ylabel("Eixo y") # Eixo y
plt.title("Gráfico 2") # titulo do gráfico
plt.legend()
plt.subplot(linhas,colunas,3)
plt.plot(Dados_x*Dados_x, 'g--', label = "Dados_x^2")
plt.xlabel("Eixo x") # Eixo x
plt.ylabel("Eixo y") # Eixo y
plt.title("Gráfico 3") # titulo do gráfico
plt.legend()
plt.subplot(linhas,colunas,4)
plt.plot(Dados_y, 'c--', label='Dados_y')
plt.xlabel("Eixo x") # Eixo x
plt.ylabel("Eixo y") # Eixo y
plt.title("Gráfico 4") # titulo do gráfico
plt.legend()
plt.show() #remove o warning
```

```python
# gráfico de barras
Figura = plt.figure() # Cria uma variável chamada figura
Grafico = Figura.add_axes([0,0,0.7,0.7]) # Cria os eixos para o gráfico
Rotulos = ['A','B','C','D','E','F','G','H','I','J'] #Cria rotulos
Grafico.bar(Rotulos,Dados_x, color='red') # Plota o grafico
```

```python
Grafico.bar(Rotulos,Dados_x, color='red') #dados 1
Grafico.bar(Rotulos,Dados_y*0.1, color='blue') # dados 2
Figura
```

```python
Grafico.bar(Dados_x,Dados_y/2,color=['red','blue']) # exibe os dados um lado do outro
Figura
```
